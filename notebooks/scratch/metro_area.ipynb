```python
import pandas as pd
import geopandas as gpd
from geosnap import DataStore
```

```python
store = DataStore()
```

```python
msas = store.msas()
```

```python
# split city and state
msas['state'] = msas['name'].apply(lambda x: x.split(',')[1])
```

```python
# split primary state from others
msas['state'] = msas.state.apply(lambda x: x.split('-')[0])
```

```python
# there's still a trailing ' '
msas['state'] = msas['state'].str[1:]
```

```python
msas.state[0]
```

```python
# missing data for PR
msas = msas[msas.state!='PR']
```

```python
msas
```

```python
regions = pd.read_csv("../../data/bea_regions.csv")
```

```python
states = store.states()
```

```python
states = states.merge(regions, left_on='name', right_on='state_name', how='left')
```

```python
states = states.rename(columns={'region':'bea_region'})
```

```python
states = states.drop(columns=['Unnamed: 0'])
```

```python
states
```

```python
state_fips = pd.read_csv('../../data/state_fips.csv', converters={'st':str})
```

```python
states = states.merge(state_fips.set_index('stname')[' stusps'], left_on='state_name', right_index=True).rename(columns={' stusps':'state_abbrev'})
```

```python
states.state_abbrev = states.state_abbrev.str[1:]
```

```python
states
```

```python
states.to_parquet("../../data/bea_region_table.parquet")
```

```python
msas = msas.merge(states.set_index('state_abbrev')['bea_region'], right_index=True, left_on='state')
```

```python
msas
```

```python
msas.to_parquet("../../data/msa_region.parquet")
```

```python
tiger = gpd.read_parquet("../../data/census/msa_population.parquet")
```

```python
tiger
```

```python
metro_msas = msas.merge(tiger.set_index('GEOID')[['ALAND', 'AWATER', 'population', 'pop_density']], left_on='geoid', right_index=True)
```

```python
metro_msas = metro_msas.rename(columns={'ALAND': 'land_area_sqkm', 'AWATER': 'water_area_sqkm'})
```

```python
# convert square meters to square kilometers
metro_msas.land_area_sqkm = metro_msas.land_area_sqkm / 1e6
metro_msas.water_area_sqkm = metro_msas.water_area_sqkm / 1e6
```

```python
metro_msas['pop_density'] = metro_msas.population / metro_msas.land_area_sqkm
```

```python
metro_msas
```

```python
metro_msas.to_parquet("../../data/metros.parquet")
```
