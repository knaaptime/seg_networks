```python
import pandas as pd
from geosnap import DataStore
from seg_networks import generate_network_measures
import os
```

```python
from tqdm.auto import tqdm
```

```python
datasets = DataStore()
```

```python
metros = datasets.msas()[datasets.msas()['type']=='Metro Area']
```

## Loop over metros and compute network topology metrics

```python
failed = []
with tqdm(total=metros.shape[0]) as t:

    for msa in metros['geoid'].astype(str).tolist():
        if not os.path.exists("../data/graphs/{msa}_graph.graphml"):
            try:
                metrics = generate_network_measures(msa)
                metrics.to_csv(f"../data/{msa}_measures.csv")
            except Exception as e:
                print(f"{msa} failed with {e}")
                failed.append(msa)
            else:
                pass
        t.update(1)
```
